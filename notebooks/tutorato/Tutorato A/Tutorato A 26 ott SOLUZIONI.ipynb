## Es. 1:
Scrivere una funzione `sum n` che calcoli **ricorsivamente** la somma di tutti i numeri da 0 a n. Usarla poi per calcolare `sum 1000000`

```ocaml
let rec sum n = 
    if n = 0 then 0
    else n + sum (n-1);;

sum 1000000;;
```

```ocaml
let sum2 n =
    let rec aux n tot =
        if n = 0 then tot
        else aux (n-1) (tot+n)
    in aux n 0;;

sum2 1000000;;
```
